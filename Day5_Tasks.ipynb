# Task 3

```python
#Find the average salary for every employee according to task1 
```

```python
data = {
    'Name': ['Akash', 'Binod', 'Chetan', 'Darshan', 'Ivaan', 'ankit', 'Geet', 'Harsh', 'Anuj', 'Jagat'],
    'Salary': [50000, 60000, 55000, 70000, 48000, 75000, 60000, 80000, 65000, 55000],
    'Age':[22,24,45,34,56,43,32,25,36,48],
    'Departments':['HR','Sales','Engineering','Medical','Marketing','HR','Sales',
                   'Engineering','Medical','Marketing']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
avg_salaries = df.groupby('Name')['Salary'].mean()
```

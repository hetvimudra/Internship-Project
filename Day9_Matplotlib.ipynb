```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,100)
```

```python
x
```

```python
y=np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
plt.title("Wave graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.bar(x,y)
```

```python
x=[10,20,30,40,50]
y=[20,30,40,50,60]
```

```python
plt.bar(x,y,color="green")
plt.title("Bar graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
y=["A","B","c","D","E"]
```

```python
plt.pie(x,labels=y,shadow=True,autopct="%0.0f",explode=(0,0,0.3,0,0.1))
plt.show()
```

```python
plt.scatter(x,y)
```

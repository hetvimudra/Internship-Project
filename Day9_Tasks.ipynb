```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:/Users/hetvi/Desktop/15-Days Internship/train.csv")
```

```python
df
```

```python
df.info()
```

```python
gender = df["Sex"].value_counts()
```

```python
# Create a pie chart
labels = gender.index
sizes = gender.values
colors = ['blue', 'pink']  # Custom colors for male and female

plt.pie(sizes, labels=labels, colors=colors, autopct='%0.0f', shadow=True)

#plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Gender Distribution in Titanic Dataset")

plt.show()
```

```python
# Plot bar graph how many alive people using survive column in train.csv dataset
```

```python
#df["Survived"]=df["Survived"].value_counts()

# Plot a bar graph
plt.bar(df["Survived"].index, df["Survived"].values, color=['red', 'green'])
plt.xlabel("Survived")
plt.ylabel("Number of Passengers")
plt.xticks([0, 1], ["Not Survived", "Survived"])  # Set x-axis labels
plt.title("Number of Passengers by Survival in Titanic Dataset")

plt.show()
```

# Task 3

```python
# plot unique value of cabin column and subplot of cabin and gender column
```

```python
unique_cabins = df["Cabin"].nunique()

# Plot unique values in the "Cabin" column
plt.figure(figsize=(8, 5))
plt.bar(["Unique Cabins"], [unique_cabins], color='purple')
plt.xlabel("Column")
plt.ylabel("Number of Unique Values")
plt.title("Number of Unique Values in Cabin Column")
plt.show()

# Create a subplot of "Cabin" and "Gender" columns
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

df["Cabin"].value_counts().plot(kind="bar", ax=axes[0], color='blue')
axes[0].set_xlabel("Cabin")
axes[0].set_ylabel("Count")
axes[0].set_title("Counts of Passengers by Cabin")

df.groupby(["Cabin", "Sex"]).size().unstack().plot(kind="bar", stacked=True, ax=axes[1])
axes[1].set_xlabel("Cabin")
axes[1].set_ylabel("Count")
axes[1].set_title("Counts of Passengers by Cabin and Gender")
axes[1].legend(title="Sex")

plt.tight_layout()
plt.show()
```
